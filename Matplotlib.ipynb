## SubPlot

```python
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y_cos,'+',color='r')

plt.subplot(2,1,1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Cos & Sin Plot")
plt.plot(x,y_sin,'*',color='orange')
plt.show()
```

```python
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Cos Plot")
plt.plot(x,y_cos,'+',color='y')

plt.subplot(2,1,2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sin Plot")
plt.plot(x,y_sin,'+',color='m')
plt.show()
```

```python
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Cos Plot")
plt.plot(x,y_cos,'+',color='y')

plt.subplot(2,1,2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sin Plot")
plt.plot(x,y_sin,'+',color='b')
plt.subplots_adjust(hspace=0.9)
plt.show()
```

```python
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.subplot(3,1,1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Cos Plot")
plt.plot(x,y_cos,'_',color='y')

plt.subplot(3,1,2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sin Plot")
plt.plot(x,y_sin,'+',color='r')

plt.subplot(3,1,3)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Tan plot")
plt.plot(x,y_tan,'>',color='g')

plt.subplots_adjust(hspace=0.9)
plt.show()
```
